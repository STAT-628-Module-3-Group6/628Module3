```python
#basic basemap of the world
plt.figure(1, figsize=(6,6))
# use ortho projection for the globe type version
m1=Basemap(projection='ortho',lat_0=20,lon_0=-50)

# hex codes from google maps color 
#add continents
m1.fillcontinents(color='#bbdaa4',lake_color='#4a80f5') 
# add the oceans
m1.drawmapboundary(fill_color='#4a80f5')                
# Draw the boundaries of the countires
m1.drawcountries(linewidth=0.1, color="black")

#Add the scatter points to indicate the locations of the businesses
mxy = m1(restBus["longitude"].tolist(), restBus["latitude"].tolist())
m1.scatter(mxy[0], mxy[1], s=3, c="red", lw=3, alpha=1, zorder=5)
plt.title("Yelp Restaurants")
plt.savefig('./plot1.jpg')
plt.show()
```

```python
df_reviewsPerCity = pd.DataFrame(restBus.groupby(['city'])['review_count'].sum().sort_values(ascending=False))
df_openPerCity=pd.DataFrame(restBus.groupby(['city'])['business_id'].count().sort_values(ascending=False))
df = pd.merge(df_reviewsPerCity, df_openPerCity, how='left', on='city')
df_reviewPercent=pd.DataFrame(restBus.groupby(['city'])['review_count'].sum().sort_values(ascending=False)/df['review_count'].sum()*100.0)
df_openPercent=pd.DataFrame(restBus.groupby(['city'])['business_id'].count().sort_values(ascending=False)/df['business_id'].sum()*100.0)
df_percent = pd.merge(df_reviewPercent, df_openPercent, how='left', on='city')
```

```python
Categories={}
for cat in restBus["categories"]:
    all_categories= cat.split(", ")
    for x in all_categories:
        try :
            Categories[x] =Categories[x]+1
        except:
            Categories[x]=1
del Categories['Restaurants'],Categories['Food'],Categories['Nightlife'],Categories['Bars'],Categories['Sandwiches'],Categories['Coffee & Tea'],Categories['Fast Food'],Categories['Pizza'],Categories['Burgers'],Categories['Breakfast & Brunch'],Categories['Bakeries'],Categories['Grocery'],Categories['Desserts'],Categories['Event Planning & Services'],Categories['Cafes'],Categories['Shopping'],Categories['Ice Cream & Frozen Yogurt']
del Categories['Chicken Wings'],Categories['Salad'],Categories['Seafood'],Categories['Sushi Bars'],Categories['Beer']
del Categories['Wine & Spirits'],Categories['Caterers'],Categories['Delis'],Categories['Convenience Stores'],Categories['Sports Bars']





```

```python
num_cat =10 # to show top 10 catrgories
top_categories = pd.DataFrame.from_dict(data= Categories,orient="index")
top_categories.reset_index(inplace=True)
top_categories.columns = ['category', 'occurance']

x_val=top_categories.sort_values("occurance")[::-1][:num_cat].occurance.values
labels=top_categories.sort_values("occurance")[::-1][:num_cat].category.values
series = pd.Series(x_val, index=labels)
series.plot.bar(figsize=(8, 6))
plt.ylabel('Number of Restaurants', fontsize=12)
plt.xlabel('Restaurants Categories', fontsize=12)
#worst=business.loc[business["stars"]==1.0].sort_values(by="review_count",ascending=False)
plt.savefig('./plot3.jpg')
```
