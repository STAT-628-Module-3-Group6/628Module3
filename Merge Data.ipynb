```python
import pandas as pd
import numpy as np
import json
```

```python
business = [json.loads(line) for line in open('Data/business.json', 'r', encoding = 'UTF-8')]
```

```python
## extract the business_id to a list
business_count = len(business)
business_id = []
i = 0
while i < business_count:
    business_id.append(business[i]["business_id"])
    i = i + 1
```

```python
### Combine the reviews and business

count = 0
review_count = [0] * business_count
with open("Data/review.json", 'r', encoding='UTF-8') as f:
    for line in f:
        one_line = json.loads(line)
        temp_index = business_id.index(one_line["business_id"]) ## the index of the related business_id in the business list
        ## print(temp_index)
        current_number_review = review_count[temp_index]

        if(current_number_review == 0):
            first_element = list()
            first_element.append(one_line)
            business[temp_index]["Review"] = first_element
        else:
            business[temp_index]["Review"].append(one_line)

        review_count[temp_index] += 1
        count += 1
        if(count % 10000 == 0):
            print(count)
    
```

```python
### Combine the Tip and business

count2 = 0
Tip_count = [0] * business_count
with open("Data/tip.json", 'r', encoding='UTF-8') as f:
    for line in f:
        one_line = json.loads(line)
        temp_index = business_id.index(one_line["business_id"]) ## the index of the related business_id in the business list
        current_number_Tip = Tip_count[temp_index]

        if(current_number_Tip == 0):
            first_element = list()
            first_element.append(one_line)
            business[temp_index]["Tip"] = first_element
        else:
            business[temp_index]["Tip"].append(one_line)

        Tip_count[temp_index] += 1
        count2 += 1
        if(count2 % 10000 == 0):
            print(count2)
    
```
